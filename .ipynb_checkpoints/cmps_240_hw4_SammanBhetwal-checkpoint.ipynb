### c)


```python
Norway_data= data.loc[data['Country'] == 'NOR']
list(Norway_data.Medal)




```

### d)

```python
Norway_year_medal = Norway_data.groupby(['Medal', 'Year'])['Medal'].count()
Norway_year_medal
```

```python
Norway_by_year = Norway_data.groupby('Year')['Medal'].count()
Norway_by_year
```

### e)

```python
highest_medal = Norway_by_year.idxmax()
print(f'Norway obtain the highest number of medals in {highest_medal}.')
```

# Question number 2

### a)

```python
Objects = pd.DataFrame({'Color': ['white', 'red', 'green', 'red',
'green','white'],
                           
            'Object':['pen', 'pencil', 'pencil', 'ashtray',
'pen','ashtray'],
                           
                'Price1':[5.56, 4.24, 2.50, 0.56, 2.83, 6.44],
                'Price2':[4.75, 4.12, 1.60, 0.75, 5.15, 4.00]})
Objects
```

### b)

```python
Objects.groupby('Object')['Price2'].sum()
```

### c)

```python
#calculating mean
Objects.groupby('Color').agg(mean= ('Price1', 'mean'))
```

```python
#calculating range

Objects.groupby('Color').max()['Price1'] - Objects.groupby('Color').min()['Price1']
```
