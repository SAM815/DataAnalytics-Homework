## Question number 1:

The indexing operator [] is generally preferred over the dot notation because the dot notation can cause confusion when a DataFrame contains a method with the same name as a column, while the indexing operator always refers to a column.  Also, the dot noteation cannot be used when the column name contains spaces or special characters that are not allowed in Python variable names. Finally, the indexing operator allows for more flexibility when selecting multiple columns or subsets of rows and columns.
